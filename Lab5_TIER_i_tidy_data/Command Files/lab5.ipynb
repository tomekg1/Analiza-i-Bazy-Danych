1. Wczytanie danych, zmiana nazw kolumn i wyświetlenie informacji

```python
import pandas as pd

df = pd.read_csv("..\\Original data\\drinks.csv")
df.rename(columns= {'beer_servings':'beer servings', 
"spirit_servings": "spirit servings", 
"wine_servings": "wine servings",
"total_litres_of_pure_alcohol": "total litres of pure alcohol"}, inplace=True)
display(df)
```

2. Kraje z najwyższym spożyciem czystego alkoholu

3. Kraje z zerowym spożyciem alkoholu

4. Usunięcie krajów z zerowym spożyciem alkoholu i zapisanie wynikowej tabeli

```python
csv_file = pd.read_csv("..\\Original data\\drinks.csv")

delete = []
for i, data in csv_file.total_litres_of_pure_alcohol.iteritems():
    if data == 0:
        delete.append(i)
        
df = csv_file.drop(delete, axis=0)

df.to_csv("../Analysis Data/final_data.csv", index=False)
```
