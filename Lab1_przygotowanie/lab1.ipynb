Wykresy funkcji dla zadanych przedziałów.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

f1 = lambda x: x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)
plt.plot(x1, f1(x1))
plt.title("f(x^2 + 5) dla przedziału (-1, 1)")
plt.xlabel("x")
plt.ylabel("f(x^2 + 5)")
plt.legend(["f(x^2 + 5)"], loc="upper right")
plt.show()
```

```python
plt.plot(x2, f1(x2))
plt.title("f(x^2 + 5) dla przedziału (-6, 6)")
plt.xlabel("x")
plt.ylabel("f(x^2 + 5)")
plt.legend(["f(x^2 + 5)"], loc="upper right")
plt.show()
```

```python
plt.plot(x3, f1(x3))
plt.title("f(x^2 + 5) dla przedziału (0, 5)")
plt.xlabel("x")
plt.ylabel("f(x^2 + 5)")
plt.legend(["f(x^2 + 5)"], loc="upper right")
plt.show()
```
