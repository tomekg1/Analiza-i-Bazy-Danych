Zadanie 2

```python
s = '''
SELECT cu.first_name, cu.last_name, ci.city 
FROM customer cu 
INNER JOIN address ad ON cu.address_id = ad.address_id
INNER JOIN city ci ON ad.city_id = ci.city_id
ORDER BY ci.city ASC
'''
df = pd.read_sql(s, conn)
display(df)
```

Zadanie 3

```python
s = '''
SELECT f.title, AVG(p.amount) FROM payment p
INNER JOIN rental r ON p.rental_id = r.rental_id
INNER JOIN inventory i ON r.inventory_id = i.inventory_id
INNER JOIN film f ON i.film_id = f.film_id
GROUP BY f.title
ORDER BY f.title ASC
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 4

```python
s = '''
SELECT c.name, COUNT(f.title) FROM category c
INNER JOIN film_category fc ON c.category_id = fc.category_id
INNER JOIN film f ON fc.film_id = f.film_id
GROUP BY c.name
ORDER BY c.name ASC
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 5

```python
s = '''
SELECT co.country, COUNT(cu.customer_id) FROM country co
INNER JOIN city ci ON co.country_id = ci.country_id
INNER JOIN address ad ON ci.city_id = ad.city_id
INNER JOIN customer cu ON ad.address_id = cu.address_id
GROUP BY co.country
ORDER BY co.country ASC
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 6

```python
s = '''
SELECT s.*, COUNT(c.customer_id) FROM store s
INNER JOIN customer c ON s.store_id = c.store_id
GROUP BY s.store_id
HAVING COUNT(c.customer_id) > 200 AND COUNT(c.customer_id) < 300
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 7

```python
s = '''
SELECT first_name, last_name, SUM(f.length) FROM customer cu
INNER JOIN rental re ON cu.customer_id = re.customer_id
INNER JOIN inventory inv ON re.inventory_id = inv.inventory_id
INNER JOIN film f ON inv.film_id = f.film_id
GROUP BY cu.first_name, cu.last_name
HAVING SUM(f.length) > 200
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 8

Zadanie 10

```python
s = '''
SELECT c.name, f.title, f.length FROM film f
INNER JOIN film_category fc ON f.film_id = fc.film_id
INNER JOIN category c ON fc.category_id = c.category_id
WHERE f.length = 
    (SELECT MAX(f.length) FROM film f
    INNER JOIN film_category fc ON f.film_id = fc.film_id
    INNER JOIN category cat ON fc.category_id = c.category_id 
    WHERE cat.name = c.name)
ORDER BY c.name, f.title ASC
'''
df = pd.read_sql(s, con=conn)
display(df)
```

Zadanie 11

```python
s = '''
SELECT c.name, f.title, f.length FROM category c
INNER JOIN film_category fc ON c.category_id = fc.category_id
INNER JOIN film f ON fc.film_id = f.film_id
WHERE f.length = 
    (SELECT MAX(f.length) FROM film f
    INNER JOIN film_category fc ON f.film_id = fc.film_id
    INNER JOIN category cat ON fc.category_id = c.category_id 
    WHERE cat.name = c.name)
GROUP BY c.name, f.title, f.length
HAVING COUNT(c.name) =1

ORDER BY c.name ASC

'''
df = pd.read_sql(s, con=conn)
display(df)
```
